```python
import pandas as pd
import json
```

```python
df = pd.read_excel('OppoMalaysia_Raw_File.xlsx')
```

```python
df.head()
```

```python
df['shares'] = df['shares'].replace('nan', "{'count': 0}")
```

```python
columnsToParse = ['reactions', 'shares', 'reactions_list', 'comments', 'interaction_detail']

for column in columnsToParse:
    df[column] = df[column].astype(str)
    df[column] = df[column].apply(lambda x:eval(x))
```

```python
df.head()
```

```python
#filtering
df['created_timestamp'] = pd.to_datetime(df['created_timestamp'])
```

```python
df.dtypes
```

```python
start_date = "2020-05-01"
end_date = "2021-04-30"

after_start_date = df["created_timestamp"] >= start_date
before_end_date = df["created_timestamp"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = df.copy().loc[between_two_dates]
```

```python
[x for x in filtered_dates['reactions_list'].iloc[0].values()]
```

```python
filtered_dates['shares_count'] = filtered_dates['shares'].apply(lambda x: x['count'])
filtered_dates['comments_summary_total_count'] = filtered_dates['comments'].apply(lambda x: x['summary']['total_count'])
filtered_dates['reactions_list_summary_total_count'] = filtered_dates['reactions_list'].apply(lambda x: sum(y for y in x.values()))
```

```python
filtered_dates['interactions'] = filtered_dates['video_views'] + filtered_dates['shares_count'] + filtered_dates['comments_summary_total_count'] + filtered_dates['reactions_list_summary_total_count']
```

```python
filtered_dates['non_video_interactions'] = filtered_dates['shares_count'] + filtered_dates['comments_summary_total_count'] + filtered_dates['reactions_list_summary_total_count']  
```
